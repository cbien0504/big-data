```python
from elasticsearch import Elasticsearch
def fetch_all_documents(index_name, es_host="http://localhost:9200"):
    try:
        es = Elasticsearch(es_host)
        if not es.ping():
            print("Unable to connect to Elasticsearch")
            return
        print(f"Connected to Elasticsearch at {es_host}")
        response = es.search(
            index=index_name,
            body={
                "query": {
                    "match_all": {}
                },
                "size": 1000  # Sá»‘ lÆ°á»£ng document muá»‘n láº¥y (tá»‘i Ä‘a 10,000 vá»›i API cÆ¡ báº£n)
            }
        )
        hits = response.get('hits', {}).get('hits', [])
        print(f"Found {len(hits)} documents in index '{index_name}':")

        for doc in hits:
            print(doc["_source"])  # In ná»™i dung cá»§a document
    except Exception as e:
        print(f"Error occurred: {e}")
fetch_all_documents("tweets")
```
